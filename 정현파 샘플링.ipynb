```python
import numpy as np
import matplotlib.pyplot as plt

Ts=0.005
tn = np.arange(0, 0.1+Ts, Ts)
xn = 20.*np.cos(2*np.pi*40 * tn - 0.4 * np.pi)

plt.subplot(221)
plt.plot(tn, xn)
plt.subplot(223)
plt.stem(tn, xn, 'r')

plt.title("Sampled sign signal (Ts = 0.0005)") # 샘플링주기 0.0005

Ts = 0.001
tn = np.arange(0, 0.1+Ts, Ts)
xn = 20.*np.cos(2*np.pi*40 * tn - 0.4 * np.pi)

plt.subplot(222)
plt.plot(tn, xn)
plt.subplot(224)
plt.stem(tn, xn, 'r')

plt.title("Sampled sign signal (Ts = 0.001)") # 샘플링주기 0.001
plt.tight_layout()
plt.show()
```
